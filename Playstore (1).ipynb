```python
# Rename the common column in only one dataset 
df1.rename(columns={'App': 'App_Name'}, inplace=True)
```

```python
#Merge datasets on the renamed column from df1 and original column in df2
merged_df = pd.merge(df1, df2, left_on='App_Name', right_on='App', how='inner')
```

```python
#Defining the fumction
def convert_into_bytes(column_name):
    if isinstance(column_name,str):
        if 'k' in column_name:
            return float(column_name.replace("k",""))*1024
        elif 'M' in column_name:
            return float(column_name.replace("M",""))*1024*1024
        elif 'Varies with device' in column_name:
            return np.nan
    return column_name


merged_df["Size"]=merged_df["Size"].apply(convert_into_bytes)
    
    
```

```python
#convert dtypes
merged_df["Reviews"]=pd.to_numeric(merged_df["Reviews"])
merged_df['Last Updated'] = pd.to_datetime(merged_df['Last Updated'], errors='coerce')

merged_df['Current Ver'] = merged_df['Current Ver'].astype(str).str.extract(r'([0-9]+\.[0-9]+|[0-9]+)')[0]
merged_df['Current Ver'] = pd.to_numeric(merged_df['Current Ver'], errors='coerce')

merged_df['Android Ver'] = merged_df['Android Ver'].astype(str).str.extract(r'([0-9]+\.?[0-9]*)')[0]
merged_df['Android Ver'] = pd.to_numeric(merged_df['Android Ver'], errors='coerce')



# Ensure 'Installs' is a string before replacing characters
merged_df['Installs'] = merged_df['Installs'].astype(str)

# Remove commas and plus signs
merged_df['Installs'] = merged_df['Installs'].str.replace(',', '').str.replace('+', '', regex=False)

# Convert to numeric, setting invalid values to NaN
merged_df['Installs'] = pd.to_numeric(merged_df['Installs'], errors='coerce')


# Ensure 'Price' is a string before processing
merged_df['Price'] = merged_df['Price'].astype(str)

# Remove '$' symbol safely
merged_df['Price'] = merged_df['Price'].str.replace('$', '', regex=False)

# Convert non-numeric values to NaN and change to float
merged_df['Price'] = pd.to_numeric(merged_df['Price'], errors='coerce')


```

```python
ratings=merged_df.groupby(["App_Name","App"])["Rating"].mean().sort_values(ascending=False)
ratings
```

```python
#step1: Why do some apps have a Rating of 5.0 while others have very low ratings?
#Distribution of Ratings
plt.figure(figsize=(8,5))
sns.histplot(ratings,bins=20,kde=True,color="blue")
plt.title("Distribution of App Ratings")
plt.xlabel("Ratings")
plt.ylabel("count")
plt.show()
```

If most of the ratings are near 4.0 - 4.5, it suggests that users generally rate apps favorably,

A high peak near 5.0 indicates that most apps receive high ratings.

If there are significant bars near 1.0 - 2.0, it shows a presence of poorly rated apps.

```python
#step2: Why do some apps have a very high number of Reviews, while others have very few?
#Reviews vs. Ratings (Scatter Plot)
plt.figure(figsize=(10, 5))
sns.scatterplot(x='Reviews', y='Rating', data=merged_df, alpha=0.5,color='blue')
plt.title("Reviews vs. Ratings")
plt.xlabel("Number of Reviews")
plt.ylabel("Rating")
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

Apps with very few reviews tend to have a wide range of ratings, from low to high.

As the number of reviews increases, the ratings converge around 4.0 to 5.0, indicating that popular apps tend to have higher ratings.

Some apps with high reviews and lower ratings might indicate controversial or widely criticized apps.

```python
#step3: Why do some apps have a Sentiment_Polarity close to 1.0 while others are near -1.0?
# Countplot of Sentiment Labels
plt.figure(figsize=(8, 5))
sns.countplot(x='Sentiment', data=merged_df,palette=["red","blue","green"])
plt.title("Sentiment Category Count")
plt.xlabel("Sentiment")
plt.ylabel("Number of Reviews")
plt.show()
```

If the Positive bar is the tallest, it means most users were satisfied with the apps.

If Negative or Neutral bars are tall, it indicates less satisfaction or mixed feedback.

```python
category_sentiment_counts = merged_df.groupby(['Category', 'Sentiment']).size().reset_index(name="Count")

# Plotting
plt.figure(figsize=(14, 7))
sns.barplot(data=category_sentiment_counts, x='Category', y='Count', hue='Sentiment', palette=["red","blue","green"])
plt.title("Sentiment Distribution by App Category")
plt.xlabel("Category")
plt.ylabel("Number of Reviews")
plt.xticks(rotation=45, ha='right')
plt.legend(title='Sentiment')
plt.tight_layout()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

Categories like GAME  may have higher Positive sentiments, indicating users are more satisfied.

Others might show more Neutral or Negative sentiments, revealing areas of concern or average experience.

```python
#Sentiment Polarity Distribution
plt.figure(figsize=(10, 5))
sns.histplot(merged_df['Sentiment_Polarity'], bins=30, kde=True, color='green')
plt.title("Sentiment Polarity Distribution")
plt.xlabel("Sentiment Polarity")
plt.ylabel("Count")
plt.show()
```

If the graph is right-skewed, it suggests that most user reviews are positive.

If the graph is left-skewed, it means more reviews are negative.

```python
#step4: Why do some apps have a Price of 0.0 while others have a high price?
#Visualization 4: Installs vs. Price
plt.figure(figsize=(10, 5))
sns.scatterplot(x='Price', y='Installs', data=merged_df, alpha=0.5)
plt.yscale("log")  # Log scale for better visualization
plt.title("Installs vs. Price")
plt.xlabel("Price (in $)")
plt.ylabel("Number of Installs (Log Scale)")
plt.show()
```

Most free apps (Price = 0) have higher installs, suggesting that users prefer free apps over paid ones.

```python
#step5: Why do certain apps have an Installs count significantly higher than others?
#Top Categories by Installs
top_categories = merged_df.groupby("Category")["Installs"].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_categories.index, y=top_categories.values,palette="viridis")
plt.xticks(rotation=45)
plt.title("Top 10 Categories by Total Installs")
plt.xlabel("Category")
plt.ylabel("Total Installs")
plt.show()
```
